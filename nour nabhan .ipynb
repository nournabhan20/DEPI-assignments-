### 1. Question: Merge and Sort Two Lists

**Description:** Given two lists of integers, merge them into one list and sort the result in ascending order. Do not use Python's built-in sort() function or other sorting functions.




```python
'''
# Example Input:
list1 = [3, 1, 4]
list2 = [2, 5, 0]

# Example Output:
[0, 1, 2, 3, 4, 5]'''
```

```python
list1 = input('enter the first list of the numbers separated by commas:').split(',')
list2 = input('enter the secomd list of the numbers separated by commas:').split(',')
merged_list = list1+list2

for x in range(len(merged_list)):
  for y in range(len(merged_list)-1):
    if int(merged_list[y]) > int(merged_list[y+1]):
      merged_list[y], merged_list[y+1] = merged_list[y+1], merged_list[y]

list = [int(num) for num in merged_list ]
print(list)
```

### 2. Question: Repeat Characters in a String

**Description:** Given a string, create a new string by repeating each character in the original string n times.

Assume n is given and always a positive integer.

Do not use if statements in your solution.


```python
'''
# Example Input:
s = "abc"
n = 3

# Example Output:
"aaabbbccc"

'''
```

```python
s = input('input string: ')
n = int(input('input repetition count (postive integer): '))
result = "".join([char*n for char in s])
print(result)
```
