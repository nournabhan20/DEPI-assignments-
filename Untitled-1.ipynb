```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
df= pd.read_csv('Titanic.csv')
df
```

```python
df=df.set_index('PassengerId') 
```

```python
df['Parch'].plot(kind='hist')
plt.title('Histogram of No.Parents on Board')
plt.xlabel('no. of parents')
```

```python
df['Fare'].plot(kind='line')
plt.title('Fare Paid')
plt.ylabel('Price in British Pound')
```

```python
df_Pclass=df.set_index('Pclass')
df_Pclass
```

```python
df_Pclass=df.groupby('Pclass').count()
```

```python
df_Pclass['Name'].plot(kind='pie')
plt.title('passenger class')
plt.ylabel('')
```

```python
df_sex=df.set_index('Sex')
df_sex
```

```python
df_scount=df.groupby('Sex').count()
df_scount
```

```python
df_scount['Name'].plot(kind='pie')
plt.title('female or male ')
plt.ylabel('')
```
